```python
#passo um: carregando o corpus

poema = """E agora, José?
A festa acabou,
a luz apagou,
o povo sumiu,
a noite esfriou,
e agora, José?
e agora, você?
você que é sem nome,
que zomba dos outros,
você que faz versos,
que ama, protesta?
e agora, José?

Está sem mulher,
está sem discurso,
está sem carinho,
já não pode beber,
já não pode fumar,
cuspir já não pode,
a noite esfriou,
o dia não veio,
o bonde não veio,
o riso não veio,
não veio a utopia
e tudo acabou
e tudo fugiu
e tudo mofou,
e agora, José?

E agora, José?
Sua doce palavra,
seu instante de febre,
sua gula e jejum,
sua biblioteca,
sua lavra de ouro,
seu terno de vidro,
sua incoerência,
seu ódio — e agora?

Com a chave na mão
quer abrir a porta,
não existe porta;
quer morrer no mar,
mas o mar secou;
quer ir para Minas,
Minas não há mais.
José, e agora?

Se você gritasse,
se você gemesse,
se você tocasse
a valsa vienense,
se você dormisse,
se você cansasse,
se você morresse...
Mas você não morre,
você é duro, José!

Sozinho no escuro
qual bicho-do-mato,
sem teogonia,
sem parede nua
para se encostar,
sem cavalo preto
que fuja a galope,
você marcha, José!
José, para onde?"""
```

```python
#passo 2: tokenizando o corpus

import nltk
from nltk import tokenize
nltk.download('punkt') #checando se tenho o tokenizador!
```

```python
#2.1: dividindo o poema em diferentes sentenças e considerando-as um "documento" cada

poema = poema.lower().split('\n')
poema
```

```python
#2.2 - aplicando o tokenizador

texto_tok = []
for verso in poema:
    tokens = tokenize.word_tokenize(verso, language='portuguese')
    texto_tok.append(tokens)

texto_tok
```

```python
#passo 3: inserindo os marcadores de início e fim de sentença - bigramas

from nltk.lm.preprocessing import pad_both_ends

ngramas = 2

texto_tok_pad = []

for verso in texto_tok:
    padded = list(pad_both_ends(verso,n=ngramas))
    texto_tok_pad.append(padded)

texto_tok_pad
```

```python
#passo 4: calculando os n-gramas (gerando os bigramas de cada sentença)

ngramas = 2

bigramas_poema = []
for verso in texto_tok_pad:
    bigramas = list(nltk.ngrams(verso,n=ngramas))
    bigramas_poema.append(bigramas)

bigramas_poema
```

```python
#para gerar o modelo de linguagem, é preciso que os n-gramas de tamanho menor também estejam definidos

from nltk.util import everygrams

tam_ngrama = 3 #vai gerar todos os n-gramas de até tamanho 3

ngramas_poema = []
for verso in texto_tok_pad:
    ngramas = list(everygrams(verso,max_len=tam_ngrama))
    ngramas_poema.append(ngramas)

ngramas_poema
```

```python
#passo 5: colocando todos os tokens do corpus em uma única lista

from nltk.lm.preprocessing import flatten

tokens = list(flatten(texto_tok_pad))
tokens
```

```python
#passo 6: definindo o vocabulário

from nltk.lm import Vocabulary
vocab = Vocabulary(tokens, unk_cutoff=1) #mostrar tokens que aparecem pelo menos uma vez
vocab.counts
```

```python
#simplificando o processo: passos 3 a 6 com só um comando
from nltk.lm.preprocessing import padded_everygram_pipeline

ngrams = 3
ngramas_pad, vocab = padded_everygram_pipeline(ngrams, texto_tok)
```

```python
#passo 7: treinando um modelo de linguagem

from nltk.lm import MLE

ngrams = 3
lm = MLE(ngrams)
lm.fit(ngramas_pad, vocab)
```

```python
#7.1: gerando trechos de texto usando o método lm.generate, que recebe como parâmetro o tamanho da sequência que desejo gerar
#e o contexto (com base em algum token)

lm.generate(5,text_seed=['você']) #gera sentenças aleatórias de tamanho 5 precedidas da palavra 'você'
#lm.generate(5,text_seed=['você'], random_seed=3) se eu não quiser que o resultado seja aleatório
```

```python
#7.2 verificando a probabilidade de uma palavra ocorrer no texto

lm.score(',')
```

```python
#7.3 verificando a probabilidade de uma palavra ocorrer após outra

lm.score('agora', context=['e'])
```
